```python
import matplotlib.pyplot as plt
import csv

x = []
d = []
u = []
p = []

x_e = []
d_e = []
u_e = []
p_e = []
with open("flow1d-200.dat", "r") as file:
    plots = csv.reader(file, delimiter=" ")
    for row in plots:
        x.append(float(row[0]))
        d.append(float(row[1]))
        u.append(float(row[2]))
        p.append(float(row[3]))

with open("flow1d-2000.dat", "r") as file:
    plots = csv.reader(file, delimiter=" ")
    for row in plots:
        x_e.append(float(row[0]))
        d_e.append(float(row[1]))
        u_e.append(float(row[2]))
        p_e.append(float(row[3]))

#plt.subplot(2,1,1)
plt.plot(x, d, "b.", label="d WENO", ms = 3)
plt.plot(x, u, "g+", label="u WENO", ms = 3)
plt.plot(x, p, "bx", label="p WENO", ms = 3)
#plt.legend()
#plt.subplot(2,1,2)

plt.plot(x_e, d_e, "r-", label="density")
plt.plot(x_e, u_e, "g-", label="velocity")
plt.plot(x_e, p_e, "b-", label="pressure")

plt.legend()
plt.xlabel(r"$x$")
plt.ylabel("Value")
plt.xlim([-0.5, 0.5])
plt.title("t = 1.8s")
plt.savefig("sod.png", dpi=300)
plt.show()
```
